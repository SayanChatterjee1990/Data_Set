```
from decimal import Decimal, getcontext
import math

# Set the precision for decimal calculations (e.g., 20 significant digits)
getcontext().prec = 20

def basic_cal():
  print(" Welcome to Your favorite basic calculator")
  print(" Choose what you need to use")
  print(" 1 for addition\n 2 for subtraction\n 3 for multiplication\n 4 for division\n")
  choice= int(input(" Enter your choice "))
  if choice==1:
    result=Decimal('0.0') # Changed to Decimal for precise additions
    print(" You have chosen addition")
    numbers= (input(" Enter all of your numbers sepearated by a , \n"))
    numbers= numbers.split(",")
    for number in numbers:
        result+=Decimal(number.strip()) # Changed float() to Decimal() and added strip()
    print(" Your result is:", result)


  elif choice==2:
    print(" You have chosen subtraction")
    number1= Decimal(input(" Enter the first number ")) # Changed to Decimal
    number2= Decimal(input(" Enter the second number ")) # Changed to Decimal
    print("Your result is:", number1-number2)


  elif choice==3:
    result=Decimal('1.0') # Changed to Decimal for precise additions
    print(" You have chosen multiplication")
    numbers= (input(" Enter all of your numbers sepearated by a , \n"))
    numbers= numbers.split(",")
    for number in numbers:
        result*=Decimal(number.strip()) # Changed float() to Decimal() and added strip()
    print(" Your result is:", result)


  elif choice==4:
    print(" You have chosen division")
    number1= Decimal(input(" Enter the first number ")) # Changed to Decimal
    number2= Decimal(input(" Enter the second number ")) # Changed to Decimal
    if number2!=0:
      print(" Do you want results in quotient, remainder format or in usual format?")
      sub_choice= int(input(" Enter your choice. 1 for quotient, remainder and 2 for usual format. "))
      if sub_choice==1:
        # Quotient and remainder are typically for integers, converting to int for this specific sub_choice
        # Note: This will truncate any decimal part before performing integer division/modulo.
        quotient= int(number1)//int(number2)
        remainder= int(number1)%int(number2)
        print(" Your Quotient is: ", quotient)
        print(" Your Remainder is: ", remainder)
      elif sub_choice==2:
        print(" Your result is: ", number1/number2)
      else:
        print(" Invalid choice")
    else:
      print(" The divisor can't be zero, choose another number as the second number \n")

  else:
    print(" Invalid choice")

def std_trig():
  print(" Choose what you need to use")
  print(" 1 for sin\n 2 for cos\n 3 for tan\n")
  choice= int(input(" Enter your choice "))
  if choice==1:
    print(" You have chosen sin")
    number= Decimal(input(" Enter the angle in degrees ")) # Changed to Decimal
    print(" Your result is: ", math.sin(math.radians(number)))

  elif choice==2:
    print(" You have chosen cos")
    number= Decimal(input(" Enter the angle in degrees ")) # Changed to Decimal
    print(" Your result is: ", math.cos(math.radians(number)))

  elif choice==3:
    print(" You have chosen tan")
    number= Decimal(input(" Enter the angle in degrees ")) # Changed to Decimal
    print(" Your result is: ", math.tan(math.radians(number)))

  else:
    print(" Invalid choice")


def inv_trig():
  print(" Choose what you need to use")
  print(" 1 for arcsin\n 2 for arccos\n 3 for arctan\n")
  choice= int(input(" Enter your choice "))
  if choice==1:
    print(" You have chosen arcsin")
    number= Decimal(input(" Enter the ratio, between -1 to +1 ")) #changed to decimal
    if number>1 or number<-1:
      print(" The ratio should be between -1 to +1")
    else:
      print(" Your result is: ", Decimal(math.degrees(math.asin(number))))

  elif choice==2:
    print(" You have chosen arccos")
    number= Decimal(input(" Enter the ratio, between -1 to +1 ")) #changed to decimal
    if number>1 or number<-1:
      print(" The ratio should be between -1 to +1")
    else:
      print(" Your result is: ", Decimal(math.degrees(math.acos(number))))

  elif choice==3:
    print(" You have chosen arctan")
    number= Decimal(input(" Enter the ratio")) #changed to decimal
    print(" Your result is: ", Decimal(math.degrees(math.atan(number))))

  else:
    print(" Invalid choice")


def hyperbolic():
  print(" Choose what you need to use")
  print(" 1 for sinh\n 2 for cosh\n 3 for tanh\n 4 for arcsinh\n 5 fir arccosh\n 6 for arctanh")
  choice= int(input(" Enter your choice "))
  if choice==1:
    print(" You have chosen sinh")
    number= Decimal(input(" Enter the number"))
    print(" Your result is: ", Decimal(math.sinh(number)))

  elif choice==2:
    print(" You have chosen cosh")
    number= Decimal(input(" Enter the number"))
    print(" Your result is: ", Decimal(math.cosh(number)))

  elif choice==3:
    print(" You have chosen tanh")
    number= Decimal(input(" Enter the number"))
    print(" Your result is: ", Decimal(math.tanh(number)))

  elif choice==4:
    print(" You have chosen arcsinh")
    number= Decimal(input(" Enter the number"))
    print(" Your result is: ", Decimal(math.asinh(number)))

  elif choice==5:
    print(" You have chosen arccosh")
    number= Decimal(input(" Enter the number"))
    print(" Your result is: ", Decimal(math.acosh(number)))

  elif choice==6:
    print(" You have chosen arctanh")
    number= Decimal(input(" Enter the number"))
    print(" Your result is: ", Decimal(math.atanh(number)))

  else:
    print(" Invalid choice")


def exp_log():
  print(" Choose what you need to use")
  print(" 1 for power of a number\n 2 for any root of any number \n 3 for logarithm\n")
  choice= int(input(" Enter your choice "))
  if choice==1:
    print(" You have chosen power of a number")
    number1= Decimal(input(" Enter the base number "))
    number2= Decimal(input(" Enter the power number "))
    print(" Your result is: ", Decimal(math.pow(number1, number2)))

  elif choice==2:
    print(" You have chosen any root of any number")
    number1= Decimal(input(" Enter the radicand number "))
    number2= Decimal(input(" Enter the index number "))
    print(" Your result is: ", Decimal(math.pow(number1, 1/number2)))

  elif choice==3:
    print(" You have chosen logarithm")
    number1= Decimal(input(" Enter the base number "))
    number2= Decimal(input(" Enter the argument number "))
    print(" Your result is: ", Decimal(math.log(number2, number1)))

  else:
    print(" Invalid choice")



def sci_cal():
  print(" Welcome to Your favorite scientific calculator")
  print(" Choose what you need to use")
  print(" 1 for Standrad Trigonometric Functions\n 2 for Inverse Trigonometric Functions\n 3 for Hyperbolic Functions\n 4 for Exponential Functions and logarithms\n")
  choice= int(input(" Enter your choice "))
  if choice==1:
    print(" You have chosen Standard Trigonometric Functions")
    std_trig()

  elif choice==2:
    print(" You have chosen Inverse Trigonometric Functions")
    inv_trig()

  elif choice==3:
    print(" You have chosen Hyperbolic Functions")
    hyperbolic()

  elif choice==4:
    print(" You have chosen Exponential Functions and logarithms")
    exp_log()

  else:
    print(" Invalid choice")

def run_calculator_session():
  print(" Welcome to Your favorite calculator")
  print(" Choose what you need to use")
  print(" 1 for basic calculator \n 2 for scientific calculator \n")

  choice= int(input(" Enter your choice "))
  if choice==1:
    print(" You have chosen basic calculator")
    basic_cal()
  elif choice==2:
    print(" You have chosen scientific calculator")
    sci_cal()
  else:
    print(" Invalid choice")

  print("\n Select 1 to continue \n Select 2 to Quit")
  sub_choice= int(input(" Enter your choice "))

  if sub_choice==2:
    print(" Thank you for using this calculator")
    return False # Signal to exit the main loop
  elif sub_choice== 1:
    return True # Signal to continue the main loop
  else:
    print(" Invalid choice for continuation menu.")
    return True # Continue even on invalid choice for continuation menu.

# Main application loop
while True:
  if not run_calculator_session():
    break
```
