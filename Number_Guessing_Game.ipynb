```
import random # Import the random module for generating random numbers

repeat = 'y' # Initialize a variable to control the game loop

# Main game loop: continues as long as the player wants to play again
while repeat.lower()=='y':
    count= 0 # Initialize guess counter for each new game
    print("Welcome to The Number Guessing Game!")
    print("You need to guess the correct number")
    print("Choose Your Difficulty Level: \n")

    # Get difficulty input from the user
    difficulty= input("Easy, Medium, or Hard: ")

    # Set game parameters based on difficulty
    if difficulty.lower() == 'easy':
      low_value=1
      high_value=200
      max_count= 12

    elif difficulty.lower() == 'medium':
      low_value=1
      high_value=500
      max_count= 10

    elif difficulty.lower() == 'hard':
      low_value=1
      high_value=1000
      max_count= 8

    else:
      print("Invalid Input!") # Inform user about invalid input
      continue # Restart the loop to ask for difficulty again

    # Generate a random secret number within the chosen range
    secret_value= random.randint(low_value,high_value)
    print(f"I'm thinking of a number between {low_value} and {high_value}. You have {max_count} attempts. \n")
    print("Can you guess it?")

    # Get the first guess from the user
    guess= int(input("Guess the number: "))

    # Guessing loop: continues until the correct number is guessed or max attempts are reached
    while guess != secret_value and count< max_count:
      if guess < secret_value:
        print("Too low! Try again. ")
        count += 1 # Increment attempt counter
        if count < max_count: # Only ask for input if attempts remain
            guess= int(input("Guess the number: "))
      elif guess > secret_value:
        print("Too high! Try again. ")
        count += 1 # Increment attempt counter
        if count < max_count: # Only ask for input if attempts remain
            guess= int(input("Guess the number: "))
      # Removed else: "Unknown Error!" as it was unreachable with current logic

    # Check game outcome after the guessing loop
    if guess == secret_value:
      count+=1 # Increment count one last time for the correct guess
      print("Congratulations! You guessed the number in ", count, " tries.\n")
    elif count == max_count: # If max attempts reached without guessing correctly
      print("Sorry, you've reached the maximum number of tries. The secret number was ", secret_value)

    # Ask the user if they want to play again
    while True: # Loop indefinitely until valid input is received
      print("Would you like to play again? (Y/N)")
      repeat = input("Type Y for Yes, N for No: ")
      if repeat.lower() == 'y':
        break # Exit the inner loop, continue main game loop
      elif repeat.lower() == 'n':
        # This will cause the main game loop condition `while repeat.lower() == 'y'` to be false,
        # effectively breaking out of the entire game after this iteration.
        break # Exit the inner loop, main loop will then terminate
      else:
        print("Invalid input! Please type 'Y' for Yes or 'N' for No.")
```
